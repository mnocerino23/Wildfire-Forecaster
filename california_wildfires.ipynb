The second dataset has information on Discovery DOY, Contained DOY, CountyIds, but it's in a different format so we will need to clear things up and make both datasets consistent before combining.

```python
wildfires2.astype({'Contained DOY': 'str', 'Discovered DOY': 'str'}).dtypes

for index, row in wildfires2.iterrows():
  date = wildfires2.at[index,'Discovered DOY']
  year_month_date = ''
  for i in (range(10)):
    year_month_date += date[i]
  #print(year_month_date)
  wildfires2.at[index,'Discovered DOY'] = year_month_date

for index, row in wildfires2.iterrows():
  date = wildfires2.at[index,'Contained DOY']
  if type(date) == str:
    year_month_date = ''
    for i in (range(10)):
      year_month_date += date[i]
    #print(year_month_date)
    wildfires2.at[index,'Contained DOY'] = year_month_date
  else:
    wildfires2.at[index,'Contained DOY'] = 'NaN'
```

```python
#Use Python's datetime library to handle the dates and create new columns to hold discovery and contained months in addition to day of the year for the two
wildfires2['Discovery Month'] = ''
wildfires2['Discovery DOY'] = ''
wildfires2['Contained Month'] = ''
wildfires2['Contained_DOY'] = ''
```

```python
import datetime
from datetime import datetime

for index, row in wildfires2.iterrows():
  date_discov = wildfires2.at[index, "Discovered DOY"]
  date_usable = datetime.strptime(date_discov, '%Y-%m-%d')

  wildfires2.at[index, "Discovery Month"] = date_usable.strftime("%b")
  wildfires2.at[index, "Discovery DOY"] = date_usable.strftime("%j")

  date_contained = wildfires2.at[index, "Contained DOY"]
  if date_contained != 'NaN':
    date_usable2 = datetime.strptime(date_contained, '%Y-%m-%d')

    wildfires2.at[index, "Contained Month"] = date_usable2.strftime("%b")
    wildfires2.at[index, "Contained_DOY"] = date_usable2.strftime("%j")
wildfires2.drop(['Contained DOY', 'Discovered DOY'], axis = 1, inplace = True)
```

```python
wildfires2.head(5)
```

```python
for index, rows in wildfires2.iterrows():
  wildfires2.at[index, 'Discovery DOY'] = int(wildfires2.at[index, 'Discovery DOY'])
  if wildfires2.at[index,'Contained_DOY'] != '':
    wildfires2.at[index, 'Contained_DOY'] = int(wildfires2.at[index, 'Contained_DOY'])

#Add a new feature for the number of days the fire burns uncontained
wildfires2['DaysBurn'] = ''

#Subtract for all non empty values of Contained_DOY to calculate the days each fire burned
for index, rows in wildfires2.iterrows():
  if type(wildfires2.at[index, 'Contained_DOY']) == type(wildfires2.at[index, 'Discovery DOY']):
    if wildfires2.at[index,'Contained_DOY'] - wildfires2.at[index,'Discovery DOY'] == 0:
      #If day contained = day started, the fire should be said to burn for 1 day (not 0)
      wildfires2.at[index, 'Contained_DOY'] = 1
    else:
      #take the absolute value to account for wildfires ending in january that start in december (would yield a negative w/o abs)
      wildfires2.at[index, 'DaysBurn'] = abs(wildfires2.at[index,'Contained_DOY'] - wildfires2.at[index,'Discovery DOY'])
    
```

```python
#Now we will sort out any inconsistencies between the county codes in the two datasets
#Create a dictionary mapping california counties to their Ids
counties = dict(zip(wildfires2['County'],wildfires2['CountyIds']))
print(counties)
```

```python
#When looking through the counties dictionary we see that Mexico, State of Oregon, and State of Nevada are included in 
#Although caltrain may have assisted in this firefighting efforts, we want to drop them from the dataset as we are exploring 
#only fires in the State of California. 
```

```python
#We use this chunk of code to remove four fires from the dataset that weren't actually in california
invalid_counties = ['Mexico', 'State of Oregon', 'State of Nevada']
print(len(wildfires2['County']))
for index, row in wildfires2.iterrows():
  if wildfires2.at[index,'County'] in invalid_counties:
    wildfires2.drop(index, inplace = True)
print(len(wildfires2['County']))
```

County codes in the first dataset are off for some reason so we have to correct them using the counties dictionary

```python
#Keep track of all the indices with null values. We will leave these alone
nulls = []
for index in range(len(wildfires['County'])):
  if type(wildfires['County'][index]) == float:
    nulls.append(index)
    
#add in missing county codes when key errors arise by referencing the internet
counties['Imperial'] = 13
counties['San Francisco'] = 38
counties['Douglas'] = 'NaN'
for index, row in wildfires.iterrows():
  if index not in nulls:
    wildfires.at[index, 'CountyIds'] = counties[wildfires.at[index,'County']]
```

We also can add on fire size class to the second dataset using the criteria provided by the database.. As described by the database the classes are as follows: 

A=greater than 0 but less than or equal to 0.25 acres
B=0.26-9.9 acres
C=10.0-99.9 acres
D=100-299 acres
E=300 to 999 acres
F=1000 to 4999 acres
G=5000+ acres

```python
wildfires2['Fire Size Rank'] = ''
for index, rows in wildfires2.iterrows():
  size_fire = wildfires2.at[index, 'AcresBurned']
  if size_fire >= 0 and size_fire <= 0.25:
    size_class = 'A'
  elif size_fire > 0.25 and size_fire <= 9.9:
    size_class = 'B'
  elif size_fire > 9.9 and size_fire <= 99.9:
    size_class = 'C'
  elif size_fire > 99.9 and size_fire <= 299:
    size_class = 'D'
  elif size_fire > 299 and size_fire <= 999:
    size_class = 'E'
  elif size_fire > 1000 and size_fire <= 4999:
    size_class = 'F'
  elif size_fire > 4999:
    size_class = 'G'
  wildfires2.at[index,'Fire Size Rank'] = size_class
```

Now, we only need to add DaysBurn, Discovery Month, Discovery DOY, Contained Month, and Contained_DOY features onto the first dataset before saving the two datasets

```python
wildfires = wildfires.astype({'Discovered DOY':'float', 'Contained DOY': 'float'})
#Use lambda function to add days burned feature to the dataframe. If Discovered and Contained DOY are the same the fire only burns for 1 day (not 0 days)
wildfires['DaysBurn'] = wildfires.apply(lambda x: (x['Contained DOY']- x['Discovered DOY']) if (x['Contained DOY'] != x['Discovered DOY']) else 1, axis =1)
```

```python
wildfires['Discovery Month'] = ''
wildfires['Contained Month'] = ''

for index, row in wildfires.iterrows():
  d = str(wildfires.at[index,'Discovered DOY'])[0:-2]
  d_usable = datetime.strptime(d, '%j')
  wildfires.at[index,'Discovery Month'] = d_usable.strftime('%b')

  c = str(wildfires.at[index,'Contained DOY'])
  #Some fires have not been contained so only perform for non-NaN fires
  if c != 'nan':
    c = c[0:-2]
    c_usable = datetime.strptime(c, '%j')
    wildfires.at[index,'Contained Month'] = c_usable.strftime('%b')
```
