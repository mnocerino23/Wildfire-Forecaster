```python
#Extract month and year from the existing measured date feature using the datetime library
import datetime
from datetime import datetime

for item in dataframes:
  item['Month'] =''
  item['Year'] = ''
  for index, row in item.iterrows():
    d = item.at[index, 'Measured Date']
    n = datetime.strptime(d, '%d-%b-%Y')
    item.at[index,'Month'] = n.strftime('%b')
    item.at[index,'Year'] = n.strftime('%Y')
```

Feature we want to extract: we want to retrieve the march snowpack, water content, and snow density for each station during each year

```python
#function takes in a dataframe and returns a dictionary storing 
#keys are the year
#values are tuples of the form: (snowpack depth, water content, density)
def get_snow_dictionary(df):
  snow_dict = {}
  df = df.loc[(df['Month'] == 'Mar') | (df['Month'] == 'Apr')]
  for index, rows in df.iterrows():
    if str(df.at[index,'Year']) not in snow_dict.keys():
      year = str(df.at[index,'Year'])

      #Caste depth and water content to floats
      depth = float(df.at[index,'Depth'])
      wc = float(df.at[index,'W.C.'])

      #convert density from a string holding percentage to its decimal equivalent 
      density = float(df.at[index,'Density'].replace("%", ""))/100
      snow_dict[year] = (depth, wc, density)

  return snow_dict
```
